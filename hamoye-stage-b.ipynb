```python
X, y = data.drop(labels = ['Appliances', 'lights', 'date'], axis = 1), data['Appliances']
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
x_columns = X.columns

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train = pd.DataFrame(scaler_x.fit_transform(X_train), columns = x_columns)
X_test = pd.DataFrame(scaler_x.transform(X_test), columns = x_columns)

y_train = pd.Series(scaler_y.fit_transform(y_train), name = 'Appliances')
y_test = pd.Series(scaler_y.transform(y_test), name = 'Appliances')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def get_error(model, X_test, y_test, loss = 'mae'):
    yhat = model.predict(X_test)
    if loss == 'mse':
        return mean_squared_error(y_test, yhat)
    elif loss == 'rmse':
        return mean_squared_error(y_test, yhat, squared = False)
    elif loss == 'rss':
        rss = np.sum(np.square(y_test - y_hat))
        return rss
    else:
        return mean_absolute_error(y_test, yhat)
```

```python
pa_reg = linear_model.PassiveAggressiveRegressor(max_iter = 10000)
```

```python
pa_reg.fit(X_train, y_train)
```

```python
pa_reg.score(X_test, y_test)
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
lin_reg.score(X_test, y_test)
```

```python
def get_weights_df(model, feat, col_name):
    '''
    Returns weight for each feature in feat
    '''
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```
